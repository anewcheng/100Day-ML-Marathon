```python
import csv

fn = 'bike_02_新北市.csv'
with open(fn, 'r', encoding='utf8') as csvfile:
    rows = csv.reader(csvfile)
    data_list = list(rows)
```

```python
import folium
import geocoder
```

```python
# 請在這邊填入你的程式
location = geocoder.osm('新北市').latlng
m = folium.Map(location=location, zoom_start=13)
for item in data_list[1:]:
    try:
        # item[2]=緯度, item[1]=經度, item[0]=名稱
        
        folium.Marker([float(item[6]), float(item[7])], popup=item[1],
                     icon=folium.Icon(color='blue', prefix='fa', icon='fa-child')).add_to(m)
        
    except Exception as e:
        print(e.args)
```
